```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

###############################################################################
# 1. DEFINING THE TRUE FUNCTIONS ON THE PLANE
###############################################################################

def true_mean(x, y):
    """
    A smooth function that acts as the 'true' mean in the bandit problem.
    For illustration, we pick something reasonably smooth like:
        mu(x, y) = sin(x) + 0.5*cos(y)
    """
    return np.sin(x) + 0.5*np.cos(y)

def true_sigma(x, y):
    """
    A smooth function that acts as the 'true' standard deviation in the bandit.
    For illustration, let it vary mildly across x, y. We keep it strictly positive.
        sigma(x, y) = 0.2 + 0.1*|sin(y)| + 0.1*|cos(x)|
    """
    return 0.2 + 0.1*abs(np.sin(y)) + 0.1*abs(np.cos(x))

def sample_reward(x, y):
    """
    Returns a single sample from N(mu, sigma^2) at point (x,y).
    """
    mu = true_mean(x, y)
    sigma = true_sigma(x, y)
    return np.random.normal(mu, sigma)

###############################################################################
# 2. MODEL DEFINITION (MEAN NETWORK, VARIANCE NETWORK)
###############################################################################
# For simplicity, define small MLP networks that take (x,y) and output a scalar.

class MeanNetwork(nn.Module):
    def __init__(self):
        super(MeanNetwork, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1)
        )
    def forward(self, x):
        return self.net(x)

class VarianceNetwork(nn.Module):
    """
    This network will approximate the variance or standard deviation.
    For stability, we can have it output log-variance, or something like that.
    Here, let's just output the variance but clamp it so that it does not go below 1e-6.
    """
    def __init__(self):
        super(VarianceNetwork, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1)
        )
    def forward(self, x):
        # we can ensure positivity with an exponential or a softplus
        return torch.clamp(torch.exp(self.net(x)), min=1e-6, max=10.0)

def train_separate_networks(num_samples=2000, lr=1e-3, plot=True):
    """
    1) Sample random points (x,y), get rewards. Train mean network to approximate mu.
    2) Freeze mean network. Then train variance network to approximate sigma^2
       using data (x,y) + (reward - mean_pred)^2 as target.
    """
    # Instantiate networks
    mean_net = MeanNetwork()
    var_net = VarianceNetwork()
    
    # Optimizers
    mean_optimizer = optim.Adam(mean_net.parameters(), lr=lr)
    var_optimizer = optim.Adam(var_net.parameters(), lr=lr)
    
    # Step 1) Generate random data to train the mean network
    # -------------------------------------------------------
    # We'll sample random (x, y) in some domain, e.g. [-3, 3] x [-3, 3].
    # Then record the reward.
    xs = np.random.uniform(-3, 3, size=num_samples)
    ys = np.random.uniform(-3, 3, size=num_samples)
    rewards = np.array([sample_reward(x, y) for x, y in zip(xs, ys)])
    
    # Convert to torch
    xy_tensor = torch.tensor(np.column_stack((xs, ys)), dtype=torch.float32)
    reward_tensor = torch.tensor(rewards, dtype=torch.float32).view(-1, 1)
    
    # Train mean network
    for epoch in range(100):  # small number of epochs for demonstration
        mean_optimizer.zero_grad()
        pred_mean = mean_net(xy_tensor)
        loss_mean = nn.MSELoss()(pred_mean, reward_tensor)
        loss_mean.backward()
        mean_optimizer.step()
    
    print(f"[MeanNet] Final Mean MSE loss = {loss_mean.item():.6f}")
    
    # Step 2) Freeze mean network and gather data to train variance network
    # --------------------------------------------------------------------
    for param in mean_net.parameters():
        param.requires_grad = False
    
    # Our target for variance = (reward - predicted_mean)^2
    with torch.no_grad():
        pred_mean_val = mean_net(xy_tensor)
    target_var = (reward_tensor - pred_mean_val)**2
    
    # Train variance network
    for epoch in range(100):
        var_optimizer.zero_grad()
        pred_var = var_net(xy_tensor)
        loss_var = nn.MSELoss()(pred_var, target_var)
        loss_var.backward()
        var_optimizer.step()
    
    print(f"[VarianceNet] Final Variance MSE loss = {loss_var.item():.6f}")

    if plot:
        grid_x = np.linspace(-3, 3, 50)
        grid_y = np.linspace(-3, 3, 50)
        XX, YY = np.meshgrid(grid_x, grid_y)
        ZZ_mean_true = true_mean(XX, YY)
        ZZ_sigma_true = true_sigma(XX, YY)
        
        # Predicted
        xy_grid = np.column_stack((XX.ravel(), YY.ravel()))
        with torch.no_grad():
            ZZ_mean_pred = mean_net(torch.tensor(xy_grid, dtype=torch.float32)).view(50, 50).numpy()
            ZZ_var_pred = var_net(torch.tensor(xy_grid, dtype=torch.float32)).view(50, 50).numpy()
        
        fig, axs = plt.subplots(2, 2, figsize=(10, 8))
        # True mean
        axs[0, 0].contourf(XX, YY, ZZ_mean_true, 50, cmap='viridis')
        axs[0, 0].set_title("True Mean")
        # Pred mean
        axs[0, 1].contourf(XX, YY, ZZ_mean_pred, 50, cmap='viridis')
        axs[0, 1].set_title("Predicted Mean")
        
        # True std
        axs[1, 0].contourf(XX, YY, ZZ_sigma_true, 50, cmap='viridis')
        axs[1, 0].set_title("True Sigma")
        # Pred var => compare with sqrt
        axs[1, 1].contourf(XX, YY, np.sqrt(ZZ_var_pred), 50, cmap='viridis')
        axs[1, 1].set_title("Predicted Sigma (sqrt of variance)")
        
        plt.tight_layout()
        plt.show()
    
    return mean_net, var_net



def train_separate_networks_nll(num_samples=2000, lr=1e-3, plot=True, epochs = 200):
    mean_net = MeanNetwork()
    
    # Let's define a variant of VarNet that outputs log_var for better stability
    class VarNetLogVar(nn.Module):
        def __init__(self):
            super(VarNetLogVar, self).__init__()
            self.net = nn.Sequential(
                nn.Linear(2, 32),
                nn.ReLU(),
                nn.Linear(32, 16),
                nn.ReLU(),
                nn.Linear(16, 1) # log_var
            )
        def forward(self, x):
            log_var = self.net(x)
            log_var = torch.clamp(log_var, min=-10, max=2)
            return log_var
    
    var_net = VarNetLogVar()
    mean_optimizer = optim.Adam(mean_net.parameters(), lr=lr)
    var_optimizer = optim.Adam(var_net.parameters(), lr=lr)
    
    xs = np.random.uniform(-3, 3, size=num_samples)
    ys = np.random.uniform(-3, 3, size=num_samples)
    rewards = np.array([sample_reward(x, y) for x, y in zip(xs, ys)])
    
    xy_tensor = torch.tensor(np.column_stack((xs, ys)), dtype=torch.float32)
    reward_tensor = torch.tensor(rewards, dtype=torch.float32).view(-1, 1)
    
    for epoch in range(epochs):
        mean_optimizer.zero_grad()
        pred_mean = mean_net(xy_tensor)
        loss_mean = nn.MSELoss()(pred_mean, reward_tensor)
        loss_mean.backward()
        mean_optimizer.step()
    print(f"[MeanNet] Final Mean MSE loss = {loss_mean.item():.6f}")
    
    # 2) Freeze Mean Net
    for param in mean_net.parameters():
        param.requires_grad = False
    
    # 3) Train Var Net with negative log-likelihood
    #    NLL = 0.5*((r - mu)^2 / var) + 0.5*log(var)
    for epoch in range(epochs):
        var_optimizer.zero_grad()
        
        with torch.no_grad():
            pred_mean_val = mean_net(xy_tensor)
        
        log_var = var_net(xy_tensor)
        var_ = torch.exp(log_var)
        
        diff_sq = (reward_tensor - pred_mean_val)**2
        nll = 0.5 * diff_sq / var_ + 0.5 * log_var
        
        loss_var = torch.mean(nll)  # average over batch
        loss_var.backward()
        var_optimizer.step()
    print(f"[VarianceNet] Final Variance NLL loss = {loss_var.item():.6f}")
    
    if plot:
        
        grid_x = np.linspace(-3, 3, 50)
        grid_y = np.linspace(-3, 3, 50)
        XX, YY = np.meshgrid(grid_x, grid_y)
        ZZ_mean_true = true_mean(XX, YY)
        ZZ_sigma_true = true_sigma(XX, YY)
        
        # Predicted
        xy_grid = np.column_stack((XX.ravel(), YY.ravel()))
        xy_grid_torch = torch.tensor(xy_grid, dtype=torch.float32)
        with torch.no_grad():
            ZZ_mean_pred = mean_net(xy_grid_torch).view(50, 50).numpy()
            ZZ_log_var_pred = var_net(xy_grid_torch).view(50, 50).numpy()
            ZZ_var_pred = np.exp(ZZ_log_var_pred)
        
        fig, axs = plt.subplots(2, 2, figsize=(10,8))
        # True mean
        axs[0, 0].contourf(XX, YY, ZZ_mean_true, 50, cmap='viridis')
        axs[0, 0].set_title("True Mean")
        # Pred mean
        axs[0, 1].contourf(XX, YY, ZZ_mean_pred, 50, cmap='viridis')
        axs[0, 1].set_title("Predicted Mean")
        
        # True sigma
        axs[1, 0].contourf(XX, YY, ZZ_sigma_true, 50, cmap='viridis')
        axs[1, 0].set_title("True Sigma")
        # Pred Sigma from var => sqrt(var)
        axs[1, 1].contourf(XX, YY, np.sqrt(ZZ_var_pred), 50, cmap='viridis')
        axs[1, 1].set_title("Predicted Sigma (sqrt of variance)")
        
        plt.tight_layout()
        plt.show()
    
    return mean_net, var_net


class FuzzyQNetwork(nn.Module):
    def __init__(self):
        super(FuzzyQNetwork, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 2)  
        )
    def forward(self, x):
        out = self.net(x)
        m = out[:, 0:1]
        log_sigma = out[:, 1:2]
        sigma = torch.clamp(torch.exp(log_sigma), min=1e-6, max=10.0)
        return m, sigma

def fuzzy_q_loss(m, sigma, alpha, r):
    # Compute the target mean
    target_mean = (alpha * m + r) / (sigma**2 * alpha**2)
    sigma = torch.clamp(sigma, min=1e-6)
    kl_div = torch.log(sigma / sigma) + (sigma**2 + (m - target_mean)**2) / (2 * sigma**2) - 0.5
    return kl_div.mean()

def train_fuzzy_q(num_samples=1000, alpha=0.1, lr=1e-3, plot=True):
    q_net = FuzzyQNetwork()
    optimizer = optim.Adam(q_net.parameters(), lr=lr)
    
    # Sample data
    xs = np.random.uniform(-3, 3, size=num_samples)
    ys = np.random.uniform(-3, 3, size=num_samples)
    rewards = np.array([sample_reward(x, y) for x, y in zip(xs, ys)])
    
    xy_tensor = torch.tensor(np.column_stack((xs, ys)), dtype=torch.float32)
    r_tensor = torch.tensor(rewards, dtype=torch.float32).view(-1, 1)
    
    for epoch in range(200):
        optimizer.zero_grad()
        m_pred, sigma_pred = q_net(xy_tensor)
        
        # fuzzy q loss
        loss = fuzzy_q_loss(m_pred, sigma_pred, alpha, r_tensor)
        loss.backward()
        optimizer.step()
        
        if (epoch+1) % 50 == 0:
            print(f"[Fuzzy Q-Learning] Epoch {epoch+1}, Loss: {loss.item():.6f}")
    
    if plot:
        
        grid_x = np.linspace(-3, 3, 50)
        grid_y = np.linspace(-3, 3, 50)
        XX, YY = np.meshgrid(grid_x, grid_y)
        xy_grid = np.column_stack((XX.ravel(), YY.ravel()))
        
        # True
        ZZ_mean_true = true_mean(XX, YY)
        ZZ_sigma_true = true_sigma(XX, YY)
        
        # Pred
        with torch.no_grad():
            xy_grid_torch = torch.tensor(xy_grid, dtype=torch.float32)
            ZZ_m_pred, ZZ_sig_pred = q_net(xy_grid_torch)
            ZZ_m_pred = ZZ_m_pred.view(50, 50).numpy()
            ZZ_sig_pred = ZZ_sig_pred.view(50, 50).numpy()
        
        fig, axs = plt.subplots(2, 2, figsize=(10, 4))
        # Compare means
        ax0 = axs[1, 0]
        c0 = ax0.contourf(XX, YY, ZZ_m_pred, 50, cmap='viridis')
        plt.colorbar(c0, ax=ax0)
        ax0.set_title("Fuzzy Q-Learning: m_pred")

        #True mean
        axs[0, 0].contourf(XX, YY, ZZ_mean_true, 50, cmap='viridis')
        axs[0, 0].set_title("True Mean")

        # True std
        axs[0, 1].contourf(XX, YY, ZZ_sigma_true, 50, cmap='viridis')
        axs[0, 1].set_title("True Sigma")
        
        # Compare sigmas
        ax1 = axs[1, 1]
        c1 = ax1.contourf(XX, YY, ZZ_sig_pred, 50, cmap='viridis')
        plt.colorbar(c1, ax=ax1)
        ax1.set_title("Fuzzy Q-Learning: sigma_pred")
        plt.tight_layout()
        plt.show()
    
    return q_net

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# ------------------------
# Possibilistic Fuzzy Q-Network
# ------------------------
class FuzzyQNetwork(nn.Module):
    def __init__(self):
        super(FuzzyQNetwork, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, 2)  # Output: [mean, log_variance]
        )

    def forward(self, x):
        out = self.net(x)
        mean = out[:, 0:1]
        log_variance = out[:, 1:2]
        variance = torch.exp(log_variance)
        return mean, variance

# ------------------------
# Loss Function: Possibilistic Loss
# ------------------------
def possibilistic_loss(mean, variance, target_mean, alpha=0.1):
    """
    Possibilistic loss function based on fuzzy Q-learning.
    """
    target_distribution = torch.exp(-((target_mean - mean) ** 2) / (2 * variance))
    predicted_distribution = torch.exp(-((mean - target_mean) ** 2) / (2 * variance))
    loss = torch.mean((predicted_distribution - target_distribution) ** 2)
    return loss

# ------------------------
# Training Loop
# ------------------------
def train_fuzzy_q_network(num_samples=2000, lr=1e-3, epochs=500):
    model = FuzzyQNetwork()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    # Sample training data
    xs = np.random.uniform(-3, 3, size=num_samples)
    ys = np.random.uniform(-3, 3, size=num_samples)
    rewards = np.array([np.sin(x) + np.cos(y) + np.random.normal(0, 0.1) for x, y in zip(xs, ys)])

    xy_tensor = torch.tensor(np.column_stack((xs, ys)), dtype=torch.float32)
    reward_tensor = torch.tensor(rewards, dtype=torch.float32).view(-1, 1)

    for epoch in range(epochs):
        optimizer.zero_grad()
        mean, variance = model(xy_tensor)
        loss = possibilistic_loss(mean, variance, reward_tensor)
        loss.backward()
        optimizer.step()

        if (epoch + 1) % 50 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.6f}")

    return model

# ------------------------
# Visualization
# ------------------------
if __name__ == "__main__":
    model = train_fuzzy_q_network()

    # Visualization
    grid_x = np.linspace(-3, 3, 50)
    grid_y = np.linspace(-3, 3, 50)
    XX, YY = np.meshgrid(grid_x, grid_y)
    xy_grid = np.column_stack((XX.ravel(), YY.ravel()))
    xy_grid_torch = torch.tensor(xy_grid, dtype=torch.float32)

    with torch.no_grad():
        pred_mean, pred_variance = model(xy_grid_torch)
        pred_mean = pred_mean.view(50, 50).numpy()
        pred_variance = pred_variance.view(50, 50).numpy()

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.contourf(XX, YY, pred_mean, 50, cmap='viridis')
    plt.title("Predicted Mean")
    plt.colorbar()

    plt.subplot(1, 2, 2)
    plt.contourf(XX, YY, np.sqrt(pred_variance), 50, cmap='viridis')
    plt.title("Predicted Standard Deviation")
    plt.colorbar()

    plt.show()
```
